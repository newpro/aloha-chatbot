## Overview

* This file reads in all dialog files, a manual corrected character files.
* This file outputs all dialogs sentence data in a single csv file. The format follows dialog1, dialog2 format.

```python
guide = guide.loc[~guide['char_id'].isin(['!Broken', '!Missing', '!Conflict'])]
guide
```

```python
all_files = set(guide.file_name.unique())
all_files
```

```python
def read_dialogs(dialog_df, char_names):
    # char_names. e.g., [[jake, j, ja], ['princess gum', 'princess bubble gum']]
    # return dialogs 2: [[dialogs of jake], [dialogs of pb]]
    # each dialog is pair of (dia1, dia2)
    assert len(char_names)
    # -- build char_map --
    char_map = {}  # key: char_name, value: pos of char in char_names. E.g., jake: 0
    _pos = -1
    for names in char_names:
        _pos += 1
        for name in names:
            char_map[name] = _pos
    # -- build dia --
    dia = []
    for _ in range(len(char_names)):
        dia.append([])
    # -- write --
    pre_line = None
    pre_char = None
    for row in dialog_df.itertuples():
#         print(row)
        _curr_char = row.char_name
        if row.type == 'context':
            pre_line, pre_char = None, None
            continue
        if row.type == 'action':
            continue
        # row type is dialog
        if (pre_char == _curr_char) or (pre_char is None) or (_curr_char not in char_map):
            pre_char, pre_line = row.char_name, row.text
            continue
        # condition: pre char is not current char, and pre char is not None
        dia[char_map[_curr_char]].append((pre_line, row.text))
        pre_char, pre_line = row.char_name, row.text
    assert len(dia) == len(char_names)
    return dia


_ad_test = pd.read_csv(BASE + 'AdventureTime.csv')
_res = read_dialogs(_ad_test, [['Jake', ], ['Princess Bubblegum', ]])
print(_res[0][:20])
print('------')
print(_res[1][:20])
```

```python
def build_char_names(df_show):
    char_ids = []
    res = []
    for char_row in df_show.itertuples():
        char_ids.append(char_row.char_id)
        _names = [char_row.dialog_label]
        if not pd.isna(char_row.alternatives):
#         if (type(char_row.alternatives) is str) and (not char_row.alternatives.isnull()):
            for n in char_row.alternatives.split('/'):
                _names.append(n.strip())
        res.append(_names)
    assert len(char_ids) == len(res)
    return char_ids, res

build_char_names(guide.loc[guide.show_id=='CSIVerse'])
```

```python
import csv

def write_all(output_file):
    with open(output_file, 'a+') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow('show_id,char_name,char_id,dia1,dia2'.split(','))
    for _filename in all_files:
        _df_chars = guide.loc[guide.file_name==_filename]
        _show_id = _df_chars.iloc[0].show_id
        print(_show_id)
        _show_media = _df_chars.iloc[0].media
        _df_dialogs = pd.read_csv(BASE + _filename)
        _ids, _names = build_char_names(_df_chars)
        all_dialogs = read_dialogs(_df_dialogs, _names)
        with open(output_file, 'a+') as csvfile:
            writer = csv.writer(csvfile)
            for ind in range(len(_ids)):
                char_id = _show_media[0] + str(_ids[ind])
                for dialog_pair in all_dialogs[ind]:
                    # write format: show_id, char_name, char_id, dia1, dia2.
                    writer.writerow([_show_id, _names[ind][0], char_id, dialog_pair[0], dialog_pair[1]])

write_all('all_dialogs.csv')
```
